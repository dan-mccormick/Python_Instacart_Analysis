```python
# importing dataset that was exported at the end of Exercise 4.7
ords_prods_merge = pd.read_pickle(os.path.join(path, 'IC_Data', 'IC_prepared_data', 'orders_products_updated.pkl'))
```

```python
# creating a data frame of the first 1M rows
df = ords_prods_merge[:1000000]
```

```python
df.head()
```

```python
# using groupby to create a pandas object
df.groupby('product_name')
```

```python
# using group by with agg to determine that average number of orders per user within each department 
df.groupby('department_id').agg({'user_order_count': ['mean']})
```

```python
# using .mean() rather than agg() to get the same results as the previous script 
df.groupby('department_id')['user_order_count'].mean()
```

```python
# using agg with mean, min, and max to to get those stats by deparment ID 
df.groupby('department_id').agg({'user_order_count': ['mean', 'min', 'max']})
```

```python
# creating a new column "max_order" that will list the total number of orders associated with each user_id
# this can then be used to create a flag for each row based on their total number of orders 
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['user_order_count'].transform(np.max)
```

```python
ords_prods_merge.head(50)
```

```python
# not actually using this, but this makes it so there is no limit on the amount of rows shown in the .head() 
pd.options.display.max_rows = None
```

```python
# using .loc with the newly created max_order column to create a flag for each row based on total number of orders
# associated with each user_id. 
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal Customer'
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular Customer'
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New Customer'                
```
