```python
df_instacart_dept = pd.read_pickle(os.path.join(path, 'IC_Data', 'IC_prepared_data', 'df_instacart_dept'))
```

```python
df_instacart_dept.head()
```

```python
parent_result = []

for value in df_instacart_dept['user_num_dependents']:
   if value >= 1:
    parent_result.append("Parent")
   else:
    parent_result.append("Not Parent")
```

```python
df_instacart_dept['parent_status'] = parent_result
```

```python
df_instacart_dept.groupby('parent_status')['user_id'].nunique()
```

```python
age_result = []

for value in df_instacart_dept['user_age']:
   if value <= 30:
    age_result.append("Young")
   elif (value > 30) & (value <= 45):
    age_result.append("Middle Aged")
   else:
    age_result.append("Older")
```

```python
df_instacart_dept['user_age_group'] = age_result
```

```python
df_instacart_dept.groupby('user_age_group')['user_id'].nunique()
```

```python
income_result = []

for value in df_instacart_dept['user_income']:
   if value <= 50000:
    income_result.append("Low Income")
   elif (value > 50000) & (value < 100000):
    income_result.append("Middle Income")
   else:
    income_result.append("High Income")
```

```python
df_instacart_dept['user_income_group'] = income_result
```

```python
df_instacart_dept.groupby('user_income_group')['user_id'].nunique()
```
