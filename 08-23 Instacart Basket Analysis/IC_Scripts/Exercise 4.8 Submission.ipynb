# 4. Creating a loyalty flag column per the directions in the reading.

```python
# I imported the data set after performing the steps from the reading, so i will drop the max_orders and loyalty flag
# columns and redo the process of adding them to the dataframe.
```

```python
ords_prods_merge = ords_prods_merge.drop(['max_order', 'loyalty_flag'], axis = 1)
```

```python
ords_prods_merge.head()
```

```python
# Creating new column "max_order" based on the maximum value in the user_order_count column for each user_id 
# (grouping by user_id)
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['user_order_count'].transform(np.max)
```

```python
ords_prods_merge.head()
```

```python
# Creating a loyalty_flag column based on the max_order column I just created in the previous step. Using 
# .loc to assin a loyalty value based on the max_order number. 
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal Customer'
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular Customer'
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New Customer'
```

```python
ords_prods_merge.head()
```

# 5. Checking basic statistics for purchased made by each loyalty group

```python
# Determining averag/min/max price of products purchased by each group of customers, Loyal, Regular, and New 
# will need to group by the loyalty flag and then take an average/mix/max of the prices column to determine this 
```

```python
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max']})
```

```python
# It is very interesting to see that the average price paid for items is highest for new customers, followed by
# regular customers, and finally loyal customers. This may be due to the fact that loyal customers are more used to 
# using instacart and seek out deals or bargains more than new customers who are becoming used to the platform. 
# Obviously driving customer growth is always a priority and this finding would reinforce that. 
# There also seem to be $99,999 price outliers in multiple places through out the data set as that is the max price
# within each loyalty group. These outliers are most likely skewing the mean higher for each group. 
```

# 6. Creating a spending flag for each user based on their average purchase price

```python
# Creating a flag for the average price each user pays for items. If the average is lower than 10 then they are 
# a low spender, if it is equal or higher than 10 then they are a high spender. 
```

```python
ords_prods_merge['user_avg_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merge.head()
```

```python
ords_prods_merge.loc[ords_prods_merge['user_avg_price'] > 10, 'spending_flag'] = 'High Spender'
ords_prods_merge.loc[ords_prods_merge['user_avg_price'] <= 10, 'spending_flag'] = 'Low Spender'
```

```python
ords_prods_merge.head()
```

```python
ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

```python
# copying the spending_flag frequency table to Excel for further analysis
spendingflag_freq = ords_prods_merge['spending_flag'].value_counts(dropna = False)
spendingflag_freq.to_clipboard()
```

# 7. Creating a frequency flag for each user based on their median days since prior order

```python
# Creating a flag based on the median values of days since prior order. If the median is higher than 20 then 
# they are a non-frequent customer. If the median is higher than 10 and lower than or equal to 20 then they are 
# a regular customer. if the median is lower than or wqual to 10, then they are a frequen customer 
ords_prods_merge['user_med_days_since_order'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
ords_prods_merge.head()
```

```python
# Creating a frequency flag based on each user's median days since prior order value. 
ords_prods_merge.loc[ords_prods_merge['user_med_days_since_order'] > 20, 'frequency_flag'] = 'Non-frequent customer'
ords_prods_merge.loc[(ords_prods_merge['user_med_days_since_order'] <= 20) & (ords_prods_merge['user_med_days_since_order'] > 10), 'frequency_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['user_med_days_since_order'] <= 10, 'frequency_flag'] = 'Frequent customer'
```

```python
ords_prods_merge.head()
```

```python
ords_prods_merge['frequency_flag'].value_counts(dropna = False)
```

```python
# Exporting frequency_flag frequency table for further analysis in Excel 
freq_table = ords_prods_merge['frequency_flag'].value_counts(dropna = False)
freq_table.to_clipboard()
```
