```python
# importing departement csv as a dataframe 
df_dep = pd.read_csv(os.path.join(path, 'IC_Data', 'IC_original_data', 'departments.csv'), index_col = False)
```

```python
# transposing df_dep dataframe from wide to long 
df_dep_t = df_dep.T
```

```python
# resetting the index now that the dataframe is in long format rather than wide 
df_dep_t.reset_index()
```

```python
# creating a new variable "new_header" and assigning it the value of what is in the first row (index 0)
# of the transposed department dataframe 
new_header = df_dep_t.iloc[0]
```

```python
new_header
```

```python
# creating a new dataframe that only includes the values from index 1 (row 2) through the end of the data frame. 
# This removes the first row from the data frame since we want to make those values the names of the columns 
df_dep_t_new = df_dep_t[1:]
```

```python
df_dep_t_new
```

```python
# using the new_header variable we created earlier we are now assigning the two values from that variable to be the 
# titles of the columns in the transposed department data frame 
df_dep_t_new.columns = new_header
```

```python
df_dep_t_new
```

```python
# creating a data dictionary using the data frame we just wrangled. This will provide a data dictionary that gives 
# whhich department is associated with each department id number 
data_dict = df_dep_t_new.to_dict('index')
```
